### 📊 Strategic Revenue & Category Analysis

1. **Category Performance Audit:** Conducted an in-depth analysis of `Total Revenue`, `Average Order Value (AOV)`, and `Transaction Volume`. **Electronics** was identified as the primary revenue driver.
2. **Business Model Profiling:** Developed a custom indicator to classify categories into **Premium** vs. **High-Effort** models. Findings indicate that **Electronics** is the sole "Premium" driver, whereas all other categories operate on a "High-Effort/High-Volume" basis.
3. **Temporal Trend Dynamics:** Performed a trend analysis on the category mix using both absolute and relative (%) monthly comparisons. Results show strong **seasonality** across the portfolio, with the Electronics category's performance closely mirroring the retail’s overall trajectory.
4. **Category Mix Stability:** Evaluated the Market Share variance month-over-month. The category mix remains remarkably stable, with variance rarely exceeding **1%**, indicating consistent consumer preferences.
5. **Growth Velocity (MoM):** Calculated Month-over-Month (MoM) revenue growth per category. Most segments exhibit a stable gross change of approximately **10%**, suggesting a healthy and predictable revenue stream.

---

```python
revenue_by_cat = (
    df.groupby('product_category')['total_value']
    .agg(total_revenue='sum', avg_order='mean', transaction_count='count')
    .sort_values('total_revenue', ascending=False)
)
print(revenue_by_cat)

with pd.option_context('display.max_columns', None):
    display(revenue_by_cat.head())
```

```python
bm_analysis = df.groupby('product_category')['total_value'].agg(['count', 'sum'])
bm_analysis = bm_analysis.sort_values(by='sum', ascending=False)
print(bm_analysis)
```

```python
bm_analysis = (
    df.groupby('product_category')['total_value']
    .agg(['count', 'sum'])
    .assign(
        transaction_pct = lambda x: (x['count'] / x['count'].sum()) * 100,
        revenue_pct = lambda x: (x['sum'] / x['sum'].sum()) * 100,
        efficiency_ratio = lambda x: x['revenue_pct'] / x['transaction_pct']
    )
    .sort_values(by='revenue_pct', ascending=False)
)

print(f"Transaction Volume vs. Revenue Share vs. Efficiency Ratio:\n{bm_analysis.round(2)}")
```

```python
transaction_month = df['transaction_date'].dt.to_period('M')

monthly_revenue = df.groupby(transaction_month)['total_value'].sum()
monthly_revenue.plot(kind='line', marker='o')
plt.title('Monthly Revenue Trend')
```

```python
monthly_electronics_trend = (
    df.query("product_category == 'Electronics'")
    .groupby(transaction_month)['total_value']
    .sum()
)

monthly_electronics_trend.plot(kind='line', marker='o')
plt.title('Monthly Revenue Trend: Electronics')
plt.ylabel('Revenue')
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
monthly_category_trends = (
    df.groupby([transaction_month, 'product_category'])['total_value']
    .sum()
    .groupby(level=0)
    .apply(lambda x: x / x.sum() * 100)
    .unstack()
)

print(f"Monthly Revenue Share by Category Mix:\n{monthly_category_trends.round(2)}")
```

```python
monthly_variance = monthly_category_trends.diff()
print(f"Month-over-Month Variance (Percentage Points):\n{monthly_variance.round(2)}")
```

```python
raw_revenue = df.groupby([transaction_month, 'product_category'])['total_value'].sum().unstack()
mom_growth = raw_revenue.pct_change() * 100

print(f"MoM Revenue Growth (%):\n{mom_growth.round(2)}")
```
